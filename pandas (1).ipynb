# creating series

What is a Series? A Pandas Series is like a column in a table.It is a one-dimensional array holding data of any type. 1-D like structure for homogenous data and size if immutable.

```python
# creating series using list
d = ['A','B','C']
ds1 = pd.Series(d,index=['X','Y','Z'])
ds1
```

```python
ds1['X']
```

```python
ds1[0]
```

```python
# creating series using numpy array
data = np.array(['A','B',245])
ds2 = pd.Series(data,index = ['X','Y','Z'])
ds2
```

```python
# creating series using dictionary
dict_1 = {'A':10,'B':20,'C':30}
ds3 = pd.Series(dict_1)
ds3
```

```python
ds4 = pd.Series(dict_1,index=[0,1,2])    # it will not change value
ds4
```

```python
ds4 = pd.Series(dict_1,index=['A','B',1])
ds4
```

```python
# index will be non unique
ds3 = pd.Series(dict_1,index = ['A','B','C','A','B'])
ds3
```

```python
# creating for scalar values
sc_series = pd.Series(20,index = [1,2,3,4,5])
sc_series
```

```python
# creating using numpy
sr = pd.Series(np.ones(10))
sr
```

```python
sr = pd.Series(np.arange(1,12))
sr
```

# creating Dataframes

Data frame is hetrogeneous i.e collection of different type of value

A DataFrame is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values,categorical data and more) in columns.

Each column in a DataFrame is a Series.

Example: Like Table: Name Rollno marks aryan 30 98 adiba 40 78 sachin 50 67

# Q What is the total number of reported cases and deaths related to covid 19

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
```

```python
total_deaths
```

# Q what is the overall death ratio

```python
death_rate = total_deaths/total_cases
```

```python
death_rate
```

```python
print(f"total death rate = {death_rate*100}%")
```

# Q What is the overall number of test conducted ? A total of 935310 tests were conducted before daily tests numbers were being reported.

```python
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()
total_tests
```

# Q what fraction of test returned a positive result?

```python
positive_rate = total_cases / total_tests
print("Fraction: ",positive_rate*100)
```
