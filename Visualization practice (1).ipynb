```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Line Charts

# Line graphs are used to track changes over short and long periods of time

```python
yield_apples = [0.895,0.91,0.919,0.926,0.929,0.931]
```

```python
plt.plot(yield_apples);
```

```python
yield_apples = [0.895,0.91,0.919,0.926,0.929,0.931]
years = [2010,2011,2012,2013,2014,2015]
```

```python
plt.plot(years,yield_apples);
```

```python
# adding features to it
plt.plot(years,yield_apples)
plt.xlabel("Years")
plt.ylabel("Yield");
```

```python
# plotting multiple lines
years = range(2000,2011)
apples = [.91,.78,.88,.70,.23,.99,.89,.56,.97,.88,.62]
mango = [.19,.68,.48,.73,.34,.69,.99,.65,.79,.78,.68]
```

```python
plt.plot(years,apples)
plt.plot(years,mango)
plt.xlabel("Years")
plt.ylabel("Yeild");
```

```python
# chart title and legend
plt.plot(years,apples)
plt.plot(years,mango)
plt.xlabel("Years")
plt.ylabel("Yeild")

plt.title("Crops Yield")
plt.legend(["Apples","Mango"]);
```

```python
# markers
plt.plot(years,apples,marker='o')
plt.plot(years,mango,marker='x')
plt.xlabel("Years")
plt.ylabel("Yeild")

plt.title("Crops Yield")
plt.legend(["Apples","Mango"]);
```

```python
plt.plot(years,apples,'s-r')
plt.plot(years,mango,'d--b')
plt.xlabel("Years")
plt.ylabel("Yeild")

plt.title("Crops Yield")
plt.legend(["Apples","Mango"]);
```

```python
# Changing the figure size
plt.figure(figsize=(12,6))
plt.plot(years,apples,'s-r')
plt.plot(years,mango,'d--b')
plt.xlabel("Years")
plt.ylabel("Yeild")

plt.title("Crops Yield")
plt.legend(["Apples","Mango"]);
```

# Improving Default Styles with seaborn

```python
sns.set_style("whitegrid")
plt.plot(years,apples,'s-r')
plt.plot(years,mango,'d--b')
plt.xlabel("Years")
plt.ylabel("Yeild")

plt.title("Crops Yield")
plt.legend(["Apples","Mango"]);
```

```python
sns.set_style("darkgrid")
plt.plot(years,apples,'s-r')
plt.plot(years,mango,'d--b')
plt.xlabel("Years")
plt.ylabel("Yeild")

plt.title("Crops Yield")
plt.legend(["Apples","Mango"]);
```

# Scatter Plot 

# Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to show how much one variable is affected by another

```python
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
sns.scatterplot(flowers_df.sepal_width,flowers_df.petal_width)
```

```python
## adding hues
sns.scatterplot(flowers_df.sepal_width,flowers_df.petal_width,hue=flowers_df.species,s=100)
```

```python
# customizing seaborn figures
plt.figure(figsize = (12,6))
plt.title("Sepal_dimensions")

sns.scatterplot(flowers_df.sepal_length,
               flowers_df.sepal_width,
               hue = flowers_df.species,s = 100);
```

# Histogram

# It is used to summarize discrete or continuous data that are measured on an interval scale

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

```python
plt.hist(flowers_df.sepal_width,bins=5);
```

# Multiple Histogram

```python
import numpy as np
```

```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width,alpha = 0.4,bins = np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha = 0.4,bins = np.arange(2,5,0.25));

plt.legend(['setosa','versicolor'])
```

# Bar Chart

# The bar graph is used to compare the items between different groups over time

```python
years = range(2000,2006)
apples = [0.35,0.6,0.9,0.8,0.65,0.8]
oranges = [0.4,0.8,0.9,0.7,0.6,0.8]
```

```python
plt.bar(years,oranges);
```

```python
# line chart and bar together

plt.bar(years,oranges)
plt.plot(years,oranges,'o--r')

plt.title("Oranges");
```

```python
# Like histograms bar can be stacked on top of one another.  we use the 
# bottom argument  to plt.bar()

plt.bar(years,apples)
plt.bar(years,oranges, bottom = apples);

plt.legend(['apples','oranges'])
```
